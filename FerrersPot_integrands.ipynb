```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tau = np.linspace(0,2,1000)
def integrand(tau, n, a, b, c):
    x,y,z, = 1,1,1
    return (1 - x**2/(a**2+tau) - y**2/(b**2+tau) - z**2/(c**2+tau))**(n)/np.sqrt((a**2+tau)*(a**2*b**2+tau)*(a**2*c**2+tau))
plt.figure(figsize=(20,7))
plt.xlim(0,2)
#plt.ylim(0,500000)
plt.plot(tau,integrand(tau,2, 1, 0.35, 0.2375), c = 'red')
#plt.plot(tau,integrand(tau,2, 1, 0.5, 0.3), c = 'navy')
```

```python
def force(tau, n, i, a, b, c):
    x,y,z = 3,4,5
    return -(x*(i==0) + y*(i==1) + z*(i==2))/(a**2*(i==0) + a**2*b**2*(i==1) + a**2*c**2*(i==2) + tau)*integrand(tau, n, a, b, c)
plt.figure(figsize=(20,7))
plt.xlim(0,2)
#plt.ylim(-0.003,0.2)
plt.plot(tau,force(tau,1,0, 1, 0.35, 0.2375), c = 'red')
plt.plot(tau,force(tau,1,1, 1, 0.35, 0.2375), c = 'navy')
```

```python
from scipy.optimize import fsolve

def func(tau):
    xp, yp, zp, a2, b2, c2 = 7,1,1,8,0.35,0.2375
    return xp**2/(a2+tau) + yp**2/(a2*b2+tau) + zp**2/(a2*c2+tau) - 1

lowerlim = fsolve(func,0)[0]
lowerlim
```
